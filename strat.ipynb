```python
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Assuming your merged DataFrame is named 'data' with columns 'Price_VIX' and 'Price_SPX'

# Calculate daily returns
data['VIX_ret'] = data['Price_VIX'].pct_change()
data['SPX_ret'] = data['Price_SPX'].pct_change()
data.dropna(inplace=True)

def objective(weights, cov_matrix):
    """Calculates the portfolio variance (to minimize)."""
    portfolio_variance = weights.T @ cov_matrix @ weights
    return portfolio_variance

def optimize_portfolio(returns):
    """Optimizes portfolio weights using Markowitz approach."""
    cov_matrix = returns.cov()
    num_assets = len(returns.columns)

    # Initial weights (equal allocation)
    initial_weights = np.array([1 / num_assets] * num_assets)

    # Define bounds for weights (between 0 and 1)
    bounds = [(0, 1) for _ in range(num_assets)]

    # Constraint: sum of weights must equal 1
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

    # Optimization using SLSQP solver
    result = minimize(
        objective,
        initial_weights,
        args=(cov_matrix,),
        method='SLSQP',
        bounds=bounds,
        constraints=constraints
    )

    return result.x

def rebalance_strategy_1(data, spx_ret, lookback_period=3):
    """Rebalance based on SPX consecutive losses."""
    if (spx_ret[-lookback_period:].min() < -0.02) or (np.all(spx_ret[-lookback_period:] < 0)):
        return [0.2, 0.8]  # Increase VIX weight
    else:
        return [0.8, 0.2]  # Increase SPX weight

def rebalance_strategy_2(data, spx_ret, lookback_period=5, volatility_threshold=0.015):
    """Rebalance based on SPX volatility."""
    spx_volatility = data['SPX_ret'][-lookback_period:].std()
    if spx_volatility > volatility_threshold:
        return [0.3, 0.7]  # Increase VIX weight if volatility is high
    else:
        return [0.7, 0.3]  # Increase SPX weight if volatility is low

# Initialize lists to store portfolio returns and weights for each strategy
portfolio_returns = {'Strategy_1': [], 'Strategy_2': [], 'SPX': []}
weights_history = {'Strategy_1': [], 'Strategy_2': []}

# Backtesting loop
for i in range(252, len(data)):  # Start after the initial lookback period
    # Select returns for the past year
    returns = data[['VIX_ret', 'SPX_ret']][i-252:i]

    # Get current SPX return
    spx_ret = data['SPX_ret'][i-252:i]

    # --- Strategy 1: Consecutive Losses ---
    weights_1 = rebalance_strategy_1(data, spx_ret)
    weights_history['Strategy_1'].append(weights_1)
    portfolio_ret_1 = weights_1 @ data[['VIX_ret', 'SPX_ret']].iloc[i]
    portfolio_returns['Strategy_1'].append(portfolio_ret_1)

    # --- Strategy 2: Volatility Based ---
    weights_2 = rebalance_strategy_2(data, spx_ret)
    weights_history['Strategy_2'].append(weights_2)
    portfolio_ret_2 = weights_2 @ data[['VIX_ret', 'SPX_ret']].iloc[i]
    portfolio_returns['Strategy_2'].append(portfolio_ret_2)

    # --- SPX Benchmark ---
    portfolio_returns['SPX'].append(data['SPX_ret'].iloc[i])

# Create DataFrame for analysis
results = pd.DataFrame(portfolio_returns, index=data.index[252:])

# Calculate cumulative returns
results['Cumulative_ret_1'] = (1 + results['Strategy_1']).cumprod()
results['Cumulative_ret_2'] = (1 + results['Strategy_2']).cumprod()
results['Cumulative_ret_SPX'] = (1 + results['SPX']).cumprod()

# Plot cumulative returns
results[['Cumulative_ret_1', 'Cumulative_ret_2', 'Cumulative_ret_SPX']].plot()
plt.title('Backtesting Results - Cumulative Returns')
plt.legend(['Strategy 1', 'Strategy 2', 'SPX'])
plt.show()
```

```python
########## FINDING ALPHA ##############
#########

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def test_predictors(alltick_norm, predictors_df):
    """
    Tests the predictive power of columns in predictors_df for next month's stock returns 
    in alltick_norm.

    Args:
        alltick_norm: DataFrame of stock prices, with tickers as columns and dates as rows.
        predictors_df: DataFrame of potential predictors, with dates as rows.

    Returns:
        DataFrame: Results of regressions for each ticker, including R-squared and coefficients.
    """

    # Calculate next month's returns for each stock
    returns_df = alltick_norm.pct_change(periods=1).shift(-1)

    results = []
    for ticker in alltick_norm.columns:
        # Prepare data for regression
        X = predictors_df.loc[returns_df.index]  # Align predictors with returns
        y = returns_df[ticker].dropna()

        # Check for sufficient data points for regression
        if len(y) < len(X.columns) + 1:
            print(f"Insufficient data for regression for ticker: {ticker}")
            continue  # Skip regression if not enough data

        # Fit linear regression model
        model = LinearRegression()
        model.fit(X, y)

        # Evaluate model performance (R-squared)
        r2 = r2_score(y, model.predict(X))

        # Store results
        results.append({
            'Ticker': ticker,
            'R-squared': r2,
            **{predictor: coef for predictor, coef in zip(X.columns, model.coef_)}
        })

    return pd.DataFrame(results)



predictors_df = 
regression_results = test_predictors(alltick_norm, predictors_df)
print(regression_results)
```

```python
def calculate_conviction(qty, delta_own, scaler = 100):
    """Calculates a conviction score based on Qty and Î”Own."""

    # Normalize Qty (e.g., divide by average Qty in your dataset)
    normalized_qty = qty / 100000  # Example normalization, adjust as needed

    # Convert Î”Own to a numerical value (handle cases like 'New', '>999%')
    if delta_own == 'New':
        delta_own_num = 4.0  # High conviction for new positions
    elif delta_own == '>999%':
        delta_own_num = 4.0  # Very high conviction for large increases
    else: delta_own_num = delta_own

    # Combine normalized Qty and delta_own_num into a conviction score
    # (You can adjust the weights and formula as needed)
    conviction_score = 4 * normalized_qty + scaler* delta_own_num

    return conviction_score/20
```

```python
inside2 = inside.copy()

def merge_rows_by_ticker_date(inside):
    """
    Merges rows in the "inside" DataFrame where the same ticker appears multiple times
    on the same date, according to the specified aggregation rules.

    Args:
        inside: DataFrame containing ticker, date, qty, Delta own, Price, and other columns.

    Returns:
        DataFrame: Merged DataFrame with aggregated rows.
    """

    # Group by ticker and date
    grouped = inside.groupby(['Ticker', 'FilingDate'])

    # Define aggregation functions for each column
    agg_funcs = {
        'Qty': 'sum',
        'Î”Own': 'mean',
        'Price': 'mean'
    }

    # Get the first value for all other columns
    first_vals = {col: 'first' for col in inside.columns if col not in agg_funcs}

    # Combine aggregation functions
    agg_funcs.update(first_vals)

    # Apply aggregation and reset index
    merged_inside = grouped.agg(agg_funcs).reset_index()

    return merged_inside

# Example usage:
# Assuming your DataFrame is called "inside"

merged_inside = merge_rows_by_ticker_date(inside2)
print(merged_inside)
```

```python
# The prediction model implemented in the article is called BAROW (Batch-update AROW for Regression). It's an extension of the AROW (Adaptive Regularization of Weights) algorithm, specifically designed for online learning in non-stationary environments like financial markets.



class BAROW:
    def __init__(self, r, d):
        """
        Initialize BAROW with hyperparameter r and feature dimension d.
        """
        self.r = r
        self.d = d
        self.mu = np.zeros(d)
        self.Sigma = np.eye(d)

    def update(self, X, Y):
        """
        Update the model parameters (mu and Sigma) based on a batch of 
        features (X) and corresponding returns (Y).
        
        Args:
            X: A numpy array of shape (K, d) where K is the batch size 
               and d is the feature dimension.
            Y: A numpy array of shape (K, 1) containing the returns.
        """
        R = self.r * X.shape[0] # As defined in the article
        
        # Calculate the updated Sigma using equation (2a)
        self.Sigma = self.Sigma - self.Sigma @ X.T @ np.linalg.inv(R * np.eye(X.shape[0]) + X @ self.Sigma @ X.T) @ X @ self.Sigma

        # Calculate the updated mu using equation (2b)
        self.mu = self.mu - self.Sigma @ X.T @ np.linalg.inv(R * np.eye(X.shape[0]) + X @ self.Sigma @ X.T) @ (X @ self.mu - Y)
        
    def predict(self, X):
        """
        Predict returns based on the current model parameters.
        
        Args:
            X: A numpy array of shape (K, d) where K is the number of 
               instances to predict and d is the feature dimension.
        
        Returns:
            A numpy array of shape (K, 1) containing the predicted returns.
        """
        return X @ self.mu
```

```python
def calculate_factors(returns, A, D):
    """
    Calculates factors (F_lt) based on past returns and matrix A.
    """

    factors = pd.DataFrame(index=returns.index)
    for l in range(A.shape[1]):
        factor_components = []  # Store each weighted past return
        for k in range(1, D + 1):
            factor_components.append(A[k-1, l] * returns.shift(k))
        factors[f'F{l+1}'] = pd.concat(factor_components, axis=1).sum(axis=1).dropna()  # Sum and drop NaNs
    
    return factors
# Example usage (assuming A is already calculated):
#factors_train = calculate_factors(X_train, A, D)
```

```python
def fit_beta(factors, returns):
    """
    Fits beta coefficients using linear regression.

    Args:
        factors: DataFrame of factors.
        returns: DataFrame of stock returns.

    Returns:
        beta: Fitted beta coefficients.
    """
    model = LinearRegression()
    model.fit(factors, returns)
    beta = model.coef_

    return beta

# Example usage:
#beta = fit_beta(factors_train, X_train) 
```

```python
def calculate_metric(A, beta, returns_test, factors_test):
    """
    Calculates the metric 

    Args:
        A: Matrix with orthonormal columns.
        beta: Fitted beta coefficients.
        returns_test: DataFrame of test set returns.
        factors_test: DataFrame of test set factors.

    Returns:
        metric: Calculated metric value.
    """

    S_predictions = factors_test @ beta.T

    orthonormal_check = np.allclose(A.T @ A, np.eye(A.shape[1]), atol=1e-10) 

    if orthonormal_check:
        metric = np.mean((returns_test * S_predictions).sum(axis=1) / np.sqrt((S_predictions**2).sum(axis=1)))
    else:
        metric = -1

    return metric

```

```python
def qrt_benchmark(X_train, X_test, D=10, F=10, N_iter=50):
    """

    Args:
        X_train: Training data (returns).
        D: Time depth parameter (default: 250).
        F: Number of factors (default: 10).
        N_iter: Number of iterations (default: 1000).

    Returns:
        best_A: Best A matrix found.
        best_beta: Best beta coefficients found.
        best_metric: Best metric achieved.
    """

    best_metric = -float('inf')
    best_A = None
    best_beta = None

    for _ in range(N_iter):
        M = np.random.normal(0, 1, size=(D, D))
        A = ortho_group.rvs(dim=D)[:, :F] # Using ortho_group for orthonormal columns

        factors_train = calculate_factors(X_train, A, D)
        beta = fit_beta(factors_train, X_train)

        factors_test = calculate_factors(X_test, A, D)  # Calculate test set factors
        metric = calculate_metric(A, beta, X_test, factors_test)

        if metric > best_metric:
            best_metric = metric
            best_A = A
            best_beta = beta

    return best_A, best_beta, best_metric

# Example usage:
best_A, best_beta, best_metric = qrt_benchmark(X_train, X_test)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import ortho_group
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Data Preparation (Improved)

def prepare_data(df, test_period=250, train_period=2500):
    """Prepares data with rolling windows for training and testing."""

    returns = df.pct_change().dropna()

    # Pre-allocate NumPy arrays with correct dimensions (more efficient)
    num_samples_train = len(returns) - train_period - test_period
    num_samples_test = test_period
    num_assets = len(df.columns)

    X_train = np.zeros((num_samples_train, train_period, num_assets))
    y_train = np.zeros((num_samples_train, num_assets))  # Next day's return
    X_test = np.zeros((num_samples_test, train_period, num_assets))
    y_test = np.zeros((num_samples_test, num_assets))  # Next day's return

    for i in range(train_period, len(returns) - test_period):  # Training data
        train_idx = i - train_period
        X_train[train_idx] = returns.iloc[i - train_period:i].values
        y_train[train_idx] = returns.iloc[i].values

    for i in range(len(returns) - test_period, len(returns)):  # Test data
        test_idx = i - (len(returns) - test_period)
        X_test[test_idx] = returns.iloc[i - train_period:i].values
        y_test[test_idx] = returns.iloc[i].values

    return X_train, y_train, X_test, y_test




# Calculate Factors (No Change needed, but clarified comments)
def calculate_factors(returns, A, D):
    """Calculates factors (F_lt) based on past returns and matrix A."""
    num_assets = returns.shape[1] # get number of assets
    T = returns.shape[0] # get number of time steps
    F = A.shape[1] # get number of features
    factors = np.zeros((T, F))

    for t in range(D,T):
      for l in range(F):
        for k in range(1, D + 1):
          factors[t,l] += A[k-1,l]*returns[t-k]

    return factors


# Fit Beta (No Change Needed)
def fit_beta(factors, returns):
    """Fits beta coefficients using linear regression."""
    model = LinearRegression()
    model.fit(factors, returns)
    beta = model.coef_
    return beta

# Calculate Metric (Revised for next-day prediction)
def calculate_metric(predictions, actual_returns):
    """Calculates the Information Coefficient (IC)."""
    ic = np.corrcoef(predictions, actual_returns)[0, 1]  # Pearson correlation
    return ic

# QRT Benchmark (Modified for next-day prediction)
def qrt_benchmark(X_train, y_train, X_test, y_test, D=10, F=10, N_iter=50):
    """QRT benchmark optimized for next-day return prediction."""

    best_metric = -float('inf')
    best_A = None
    best_beta = None

    for _ in range(N_iter):
        A = ortho_group.rvs(dim=D)[:, :F] 

        factors_train = calculate_factors(X_train, A, D)
        beta = fit_beta(factors_train[D:], y_train[D:]) # Fit beta on data after the initial D periods

        factors_test = calculate_factors(X_test, A, D)
        predictions = factors_test[D:] @ beta.T  # Predict next day returns

        metric = calculate_metric(predictions.flatten(), y_test[D:].flatten())

        if metric > best_metric:
            best_metric = metric
            best_A = A
            best_beta = beta

    return best_A, best_beta, best_metric


# Autoencoder Model (Improved and integrated)
def autoencoder_model(X_train, y_train, X_test, y_test, D=10, F=10, epochs=100, batch_size=32):

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train.reshape(-1, X_train.shape[-1])).reshape(X_train.shape)
    X_test_scaled = scaler.transform(X_test.reshape(-1, X_test.shape[-1])).reshape(X_test.shape)


    input_dim = X_train.shape[2]  # Number of stocks
    timesteps = X_train.shape[1]
    
    # Encoder 
    encoder_input = Input(shape=(timesteps, input_dim))
    encoded = tf.keras.layers.LSTM(128, activation='relu')(encoder_input) # LSTM layer to capture time dependencies
    encoder_output = Dense(F, activation='linear')(encoded)

    # Decoder
    decoder_input = Input(shape=(F,))
    decoded = Dense(128, activation='relu')(decoder_input)
    decoder_output = Dense(input_dim, activation='linear')(decoded) # Output layer size matches input dimension

    # Create the encoder and decoder models
    encoder = Model(encoder_input, encoder_output, name='encoder')
    decoder = Model(decoder_input, decoder_output, name='decoder')

    # Create the autoencoder model
    autoencoder = Model(encoder_input, decoder(encoder_output), name='autoencoder')

    # Compile the model
    autoencoder.compile(optimizer='adam', loss='mse')

    # Train the model
    autoencoder.fit(X_train_scaled, X_train_scaled[:,-1,:], epochs=epochs, batch_size=batch_size) # Train to reconstruct the last return in the sequence

    # Extract the encoded representation (factors)
    factors_train = encoder.predict(X_train_scaled)
    factors_test = encoder.predict(X_test_scaled)



    # Fit beta  (using linear regression on the learned factors)
    beta = fit_beta(factors_train, y_train)

    # Predict and evaluate
    predictions = factors_test @ beta.T
    metric = calculate_metric(predictions.flatten(), y_test.flatten()) # Use revised metric for next-day prediction
    A = None # Autoencoder doesn't directly provide an 'A' matrix

    return A, beta, metric, autoencoder


# Example Usage (Demonstrating both benchmarks)
# Load your price data into a pandas DataFrame 'data_prices' with dates as index and tickers as columns
# df = data_prices.copy()



X_train, y_train, X_test, y_test = prepare_data(data_prices)
D = 20  # Time depth
F = 10  # Number of factors
N_iter = 50


# Evaluate QRT Benchmark
best_A_qrt, best_beta_qrt, best_metric_qrt = qrt_benchmark(X_train, y_train, X_test, y_test, D, F, N_iter)
print("QRT Benchmark - Best IC:", best_metric_qrt)


# Evaluate Autoencoder Model
best_A_ae, best_beta_ae, best_metric_ae, autoencoder = autoencoder_model(X_train, y_train, X_test, y_test, D, F)
print("Autoencoder Model - Best IC:", best_metric_ae)
```

```python
####### LEARN A and beta 

import numpy as np
from pymanopt.manifolds import Stiefel
from pymanopt import Problem
from pymanopt.solvers import SteepestDescent, ConjugateGradient, TrustRegions


def cost_function(A, X_train, X_test, D, F, beta):
    """
    Cost function to be minimized on the Stiefel manifold.

    Args:
        A: Matrix on the Stiefel manifold (orthonormal columns).
        X_train: Training data (returns).
        X_test: Testing data (returns).
        D: Time depth parameter.
        F: Number of factors.
        beta: Beta coefficients.

    Returns:
        cost: The negative of the metric (to be minimized).
    """

    factors_train = calculate_factors(X_train, A, D)
    factors_test = calculate_factors(X_test, A, D)

    metric = calculate_metric(A, beta, X_test, factors_test)
    return -metric # Minimize the negative of the metric


def optimize_A_stiefel(X_train, X_test, D, F, beta_init, solver_type="SteepestDescent"):
    """
    Optimizes A on the Stiefel manifold using the specified solver.

    Args:
        X_train: Training data (returns).
        X_test: Testing data (returns).
        D: Time depth parameter.
        F: Number of factors.
        beta_init: Initial beta coefficients.
        solver_type: Type of solver to use (default: "SteepestDescent").
                            Options: "SteepestDescent", "ConjugateGradient", "TrustRegions"

    Returns:
        optimized_A: Optimized A matrix.
        optimized_beta: Optimized beta coefficients.
    """

    # Define the manifold
    manifold = Stiefel(D, F)

    # Define the cost function (using a lambda to pass other parameters)
    cost = lambda A: cost_function(A, X_train, X_test, D, F, beta_init)

    # Define the problem
    problem = Problem(manifold=manifold, cost=cost)

    # Choose the solver
    if solver_type == "SteepestDescent":
        solver = SteepestDescent()
    elif solver_type == "ConjugateGradient":
        solver = ConjugateGradient()
    elif solver_type == "TrustRegions":
        solver = TrustRegions()
    else:
        raise ValueError("Invalid solver_type. Choose from 'SteepestDescent', 'ConjugateGradient', or 'TrustRegions'")

    # Solve the problem
    optimized_A = solver.solve(problem)

    # Re-fit beta with the optimized A
    factors_train = calculate_factors(X_train, optimized_A, D)
    optimized_beta = fit_beta(factors_train, X_train)

    return optimized_A, optimized_beta

# Example usage:
D = 10  # Time depth
F = 10   # Number of factors
beta_init = np.random.rand(X_train.shape[1], F) # Initialize beta randomly

# Optimize A using Riemannian Gradient Descent
optimized_A_steepest, optimized_beta_steepest = optimize_A_stiefel(X_train, X_test, D, F, beta_init, solver_type="SteepestDescent")

# Optimize A using Conjugate Gradient
optimized_A_conjugate, optimized_beta_conjugate = optimize_A_stiefel(X_train, X_test, D, F, beta_init, solver_type="ConjugateGradient")

```

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import StandardScaler


def calculate_factors_no_lookahead(X, A, D, window=60):  # Introduce a rolling window
    """
    Calculates factors using a rolling window approach to avoid lookahead bias.
    """
    num_factors = A.shape[1]
    all_dates = X.index
    factors = pd.DataFrame(index=all_dates, columns=range(num_factors))

    for i in range(window, len(all_dates)):
        current_date = all_dates[i]
        past_window = X.iloc[i - window:i]  # Data from the past 'window' days
        _, factors_window = solve_quadprog(past_window, A, D)
        factors.loc[current_date] = factors_window[-1, :] # Take the last value
    return factors.dropna() #Drop the initial nan values


def fit_beta(factors, returns):
    """Fits beta coefficients using linear regression."""
    beta = LinearRegression().fit(factors, returns).coef_
    return beta


def solve_quadprog(X, A, D=None):
    if D is None:
        D = np.identity(A.shape[0])

    # Calculate factors for each time period (each row in X)
    factors = np.linalg.lstsq(A, X.T, rcond=None)[0].T
    return None, factors

def trading_strategy_no_lookahead(A, X_train, X_test, window=60, top_n=5, short_n=5, rebalance_freq='M'):
    """Implements a long-short trading strategy without lookahead bias."""

    X = pd.concat([X_train, X_test])
    X.index = pd.to_datetime(X.index)

    portfolio_returns = pd.DataFrame(index=X_test.index, columns=['Return'])
    D = np.identity(A.shape[0])
    train_end = X_train.index[-1]

    for i, current_date in enumerate(X_test.index):

        # Start calculations from the first day of the test set
        if current_date <= train_end:
            continue

        # Recalculate factors and beta at the beginning, at rebalance frequency, or if there's a date gap
        lookback_start = max(X.index[0], current_date - pd.DateOffset(months=window))  # Efficient lookback handling
        past_window_returns = X.loc[lookback_start:current_date]

        if (i == 0) or (i > 0 and (current_date - X_test.index[i-1]).days > 1) or (
                current_date.month != X_test.index[i-1].month and rebalance_freq == 'M'):
            _, past_factors = solve_quadprog(past_window_returns, A, D)
            beta = fit_beta(pd.DataFrame(past_factors), past_window_returns)

        # Current date factors: Use rolling window UP TO the current date
        _, current_factor = solve_quadprog(X.loc[[current_date]], A)  # Use .loc[[current_date]] for single row

        predicted_returns = pd.Series(current_factor @ beta.T, index=X.columns)

        # Rank stocks
        ranked_stocks = predicted_returns.sort_values(ascending=False)

        # Long/short positions
        long_positions = ranked_stocks.index[:top_n]
        long_weights = 1 / top_n

        short_positions = ranked_stocks.index[-short_n:]
        short_weights = -1 / short_n

        # Calculate daily portfolio return
        daily_return = (X_test.loc[current_date, long_positions] * long_weights).sum() + \
                       (X_test.loc[current_date, short_positions] * short_weights).sum()  # Use X_test here
        portfolio_returns.loc[current_date, 'Return'] = daily_return

    return portfolio_returns.dropna()

portfolio_returns = trading_strategy_no_lookahead(A,X_train, X_test, window=60) 

# --- Evaluate the strategy ---
cumulative_returns = (1 + portfolio_returns).cumprod()
print("Cumulative Portfolio Returns:")
print(cumulative_returns)

# Calculate Sharpe Ratio (assuming a risk-free rate of 0 for simplicity)
sharpe_ratio = np.sqrt(252) * portfolio_returns['Return'].mean() / portfolio_returns['Return'].std()
print("Sharpe Ratio:", sharpe_ratio)
```

```python
import numpy as np
from scipy.stats import norm
from scipy.optimize import brentq

def black_scholes_call(S, K, T, r, sigma):
    """Calculates the Black-Scholes call option price."""
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

def implied_volatility_call(C, S, K, T, r):
    """Calculates the implied volatility of a call option."""

    def objective_function(sigma):
        return black_scholes_call(S, K, T, r, sigma) - C

    # Use Brent's method for robust root finding (better than Newton-Raphson in this case)
    try:
        iv = brentq(objective_function, 1e-6, 10) # Search between 0.000001 and 1000%
        return iv
    except ValueError:
        return np.nan  # Return NaN if no solution is found (e.g., bad market data)



# Example usage with simulated call prices:
S = 100  # Current stock price
K = 105  # Strike price
T = 0.5  # Time to maturity (in years)
r = 0.05  # Risk-free interest rate

# Simulate some realistic call prices with different volatilities:
true_vols = [0.2, 0.3, 0.4]  # True volatilities
call_prices = [black_scholes_call(S, K, T, r, vol) for vol in true_vols]

# Calculate implied volatilities from the simulated prices:
implied_vols = [implied_volatility_call(C, S, K, T, r) for C in call_prices]

# Print results:
for true_vol, call_price, implied_vol in zip(true_vols, call_prices, implied_vols):
    print(f"True Volatility: {true_vol:.4f}, Call Price: {call_price:.4f}, Implied Volatility: {implied_vol:.4f}")



# Example with potentially problematic data (out-of-the-money call with very low price):
C_low = 0.0001  # Very low call price â€“ might lead to no solution
iv_low = implied_volatility_call(C_low, S, K, T, r)
print(f"Low Call Price: {C_low:.4f}, Implied Volatility: {iv_low}")  # Might print NaN
```

```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.optimize import newton
import matplotlib.pyplot as plt


def implied_volatility(market_price, S, K, T, r):
    """Calculates implied volatility using Newton-Raphson."""

    def black_scholes_call(sigma):  # BS function nested for the solver
         d1 = (np.log(S / K) + (r + sigma**2 / 2) * T) / (sigma * np.sqrt(T))
         d2 = d1 - sigma * np.sqrt(T)
         return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

    def vega(sigma):  # vega for the solver
        d1 = (np.log(S / K) + (r + sigma**2 / 2) * T) / (sigma * np.sqrt(T))
        return S * np.sqrt(T) * norm.pdf(d1)

    # Use Newton-Raphson to find implied volatility
    f = lambda sigma: black_scholes_call(sigma) - market_price
    f_prime = lambda sigma: vega(sigma)

    initial_guess = 0.2  # reasonable initial guess
    implied_vol = newton(f, initial_guess, fprime=f_prime, tol=1e-6, maxiter=100)

    return implied_vol

# Example usage
market_price = 8.2
S = stock_prices[-1]  # Get the latest stock price
# K, T, r (from your code)

implied_vol = implied_volatility(market_price, S, K, T, r)
print(f"Implied Volatility: {implied_vol:.4f}")
```
