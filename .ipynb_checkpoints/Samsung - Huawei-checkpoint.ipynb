```python
import pandas as pd
import pandas_datareader as web
import datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
start = pd.to_datetime('2000-01-01')
end = pd.to_datetime('today')

df_samsung = web.DataReader('005930.KS', 'yahoo', start , end)
```

```python
# Visualisation des prix de fermeture
plt.figure(figsize=(16,8))
plt.title('Historique des prix de femeture pour Samsung', fontsize=20)
plt.plot(df_samsung['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Prix de fermeture ($)', fontsize=18)
plt.show()
```
